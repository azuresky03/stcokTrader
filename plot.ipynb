```python
import pandas as pd
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("./predictions/万华化学_pred.csv", sep=',')
df2 = pd.read_csv("./data/Test万华化学.csv", sep=',')
```

```python
figure, axes = plt.subplots(figsize=(15, 6))
axes.xaxis_date()


actuals = np.array(df2["close"].values)
dates = np.array(pd.date_range(start="2020-01-04", periods=len(actuals), freq="B"))
predictions = np.array(df1[-len(actuals):]["close"].values)

axes.plot(dates, actuals, color = 'red', label = 'Real Stock Price')
axes.plot(dates, predictions, color = 'blue', label = 'Predicted Stock Price')
plt.title('Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
```
